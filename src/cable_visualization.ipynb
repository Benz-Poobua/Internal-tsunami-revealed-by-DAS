```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path
```

```python
# define path
root = Path.cwd().parent
cable_path = os.path.join(root, 'data','cable_interp_cha.csv')
cable_path_se = os.path.join(root, 'data','cable_interp_cha_se.csv')
cable_path_check = os.path.join(root, 'data','cable_interp_cha_check.csv')
cable_aerial_path =  os.path.join(root, 'Figures','cable.png')
```

```python
# read data
cable = pd.read_csv(cable_path)
cable_se = pd.read_csv(cable_path_se)
cable_check = pd.read_csv(cable_path_check)
cable_aerial = mpimg.imread(cable_aerial_path)
```

```python
dist = cable['OD [m]']
depth = cable['Elevation [m]']

dist_check = cable_check['OD [m]']
depth_check = cable_check['Elevation [m]']

dist_se = cable_se['OD [m]']
depth_se = cable_se['Elevation [m]']
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 12), gridspec_kw={'height_ratios': [2, 1]})
ax[0].imshow(cable_aerial)
ax[0].set_title('Aerial Photo', fontsize=14)
ax[0].axis('off')

ax[1].plot(dist, depth, marker='o', label='Cross section', color='yellow')
ax[1].plot(dist_se, depth_se, marker='o', label='Cross section', color='blue')
ax[1].scatter(dist_check, depth_check, marker='o', label='Cross section', color='red', zorder=4)
ax[1].set_xlabel("OD (m)", fontsize=12)
ax[1].set_ylabel("Elevation (m)", fontsize=12)
ax[1].set_title("Cable Cross Section", fontsize=14)
ax[1].grid(True, linestyle='--', alpha=0.7)
ax[1].invert_xaxis()
# ax[1].legend()

plt.tight_layout()
plt.show()
```
